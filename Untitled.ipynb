# modeling

## create data set

Pull out all called strikes and balls for right handed batters (just to keep things simple for now).

```python
d = data[
    ((data.description == 'called_strike') | (data.type == 'B'))
    & (data.stand == 'R')
][['plate_x', 'plate_z', 'sz_top', 'sz_bot', 'type']]
```

```python
import numpy as np
observed_call = np.where(d['type'] == 'S', 1.0, 0.0)
```

Create additional features. We need locations from center of plate for the strike zone in feet to compare with `plate_x`. Also, we'll model the conditional probability of a strike given it's type (inside, high-inside, etc.).

```python
d['sz_inside'] = -8.5/12
d['sz_outside'] = 8.5 /12
```

```python
# fundamental theorem of arithmetic
vals = {
    'inside': 2,
    'outside': 3,
    'high': 5,
    'low': 7,
}
vals['inside_high'] = vals['inside'] * vals['high']
vals['inside_low'] = vals['inside'] * vals['low']
vals['outside_high'] = vals['outside'] * vals['high']
vals['outside_low'] = vals['outside'] * vals['low']
def add_pitch_type(d):
    inside = d.plate_x < d.sz_inside
    outside = d.plate_x > d.sz_outside
    high = d.plate_z > d.sz_top
    low = d.plate_z < d.sz_bot
    d['pitch_type'] = np.where(inside, 2, 1)
    d['pitch_type'] *= np.where(outside, 3, 1)
    d['pitch_type'] *= np.where(high, 5, 1)
    d['pitch_type'] *= np.where(low, 7, 1)
    d['pitch_type_desc'] = d.pitch_type.map(vals)
add_pitch_type(d)
```

```python
import pymc3 as pm
import theano.tensor as T

p_lookup = {
    'inside': 0,
    'inside_high': 1,
    'high': 2,
    'outside_high': 3,
    'outside': 4,
    'outside_low': 5,
    'low': 6,
    'inside_low': 7,
    'strike': 8
}

def is_strike(x, z,
              sz_inside, sz_outside, sz_bottom, sz_top,
              umpire_bias, p_correct):
    is_inside = x < sz_inside - umpire_bias[0]
    is_outside = x > sz_outside + umpire_bias[1]
    is_high = z > sz_top + umpire_bias[2]
    is_low = z < sz_bottom - umpire_bias[2]
    is_inside_high = is_inside & is_high
    is_outside_high = is_outside & is_high
    is_inside_low = is_inside & is_low
    is_outside_low = is_outside & is_low

    return T.switch(is_inside_high, 1-p_correct[p_lookup['inside_high']],
            T.switch(is_outside_high, 1-p_correct[p_lookup['outside_high']],
            T.switch(is_outside_low, 1-p_correct[p_lookup['outside_low']],
            T.switch(is_inside_low, 1-p_correct[p_lookup['inside_low']],
            T.switch(is_inside, 1-p_correct[p_lookup['inside']],
            T.switch(is_high, 1-p_correct[p_lookup['high']],
            T.switch(is_outside, 1-p_correct[p_lookup['outside']],
            T.switch(is_low, 1-p_correct[p_lookup['low']],
            p_correct[p_lookup['strike']]))))))))

with pm.Model() as model:
    # Define tensors for our observed data
    x = T.as_tensor(d['plate_x'])
    z = T.as_tensor(d['plate_z'])
    sz_inside = T.as_tensor(d['sz_inside'])
    sz_outside = T.as_tensor(d['sz_outside'])
    sz_bottom = T.as_tensor(d['sz_bot'])
    sz_top = T.as_tensor(d['sz_top'])
    
    # Define one distribution for each outside of strike zone location as well
    # as a distribution for inside the strike zone.
    # Priors assume the umpires usually make the correct call
    alpha_prior = pm.Uniform('alpha_prior', 2, 6)
    beta_prior = pm.Uniform('beta_prior', 2, 6)
    correct_call_probabilities = pm.Beta('correct_call_probabilities', alpha=alpha_prior, beta=beta_prior, shape=9)

    # Define distributions for the umpires bias inside, outside, top and bottom
    umpire_bias = pm.Normal('umpire_bias', 0, 5/12, shape=4)

    # Define our distribution of interest, probability of a strike.
    # Use of pm.Deterministic conditions the probability on the data.
    p_strike = pm.Deterministic('p_strike',
                                is_strike(x, z,
                                          sz_inside, sz_outside, sz_bottom, sz_top,
                                          umpire_bias,
                                          correct_call_probabilities))
    called_strike = pm.Bernoulli('called_strike', p=p_strike, observed=observed_call)
```

```python
with model:
    trace = pm.sample(5000, step=pm.Metropolis())
```
